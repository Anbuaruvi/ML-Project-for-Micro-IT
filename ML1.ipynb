```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("internship.csv")

# Check data
print(df.info())
print(df['internship_title'].head())

# Define a function to extract domain from internship title
def extract_domain(title):
    title = title.lower()
    if "data" in title and "science" in title:
        return "Data Science"
    elif "web" in title:
        return "Web Development"
    elif "machine learning" in title or "ml" in title:
        return "Machine Learning"
    elif "app" in title:
        return "App Development"
    elif "design" in title or "ui" in title:
        return "UI/UX Design"
    elif "marketing" in title:
        return "Marketing"
    elif "finance" in title:
        return "Finance"
    elif "content" in title:
        return "Content Writing"
    elif "cyber" in title:
        return "Cybersecurity"
    else:
        return "Other"

# Apply domain extraction
df['internship_domain'] = df['internship_title'].apply(extract_domain)

# Count internship domains
domain_counts = df['internship_domain'].value_counts()

# Plot top domains
plt.figure(figsize=(10,6))
sns.barplot(x=domain_counts.head(10).values, y=domain_counts.head(10).index, palette="viridis")
plt.title("Top 10 Internship Domains")
plt.xlabel("Number of Internships")
plt.ylabel("Internship Domain")
plt.tight_layout()
plt.show()

```
